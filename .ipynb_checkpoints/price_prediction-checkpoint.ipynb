```python
# outlier treatment
def remove_outliers(df):
    df.sort_values(by='BHK_NO.', inplace=True)
    q1 = np.quantile(df['BHK_NO.'], 0.25)
    q3 = np.quantile(df['BHK_NO.'], 0.75)
    IQR = q3-q1
    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    df = df[(df['BHK_NO.']>=lower_bound) & (df['BHK_NO.']<=upper_bound)]
    df.reset_index(drop=True, inplace=True)
    return df

df = remove_outliers(df)
df
```

```python
# numerical_col = []
# for col in df.columns:
numerical_col = [col for col in df.columns if df[col].dtype in ['int64', 'float64']] 
```

```python
df.corr(numeric_only = True)
```

```python
sns.heatmap(df.corr(numeric_only = True), annot=True, fmt='.1g', cmap='plasma')
plt.show()
```

```python
#  Divide X and Y (training_data)
x = df[numerical_col]
y = df[['TARGET(PRICE_IN_LACS)']]
```

```python
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=22)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# scaling 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_train
```

```python
x_test = scaler.transform(x_test)
x_test
```

```python
# train the model.
model = LinearRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
r2_score(y_test, y_pred)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```
